## Calculando uma divisão utilizando o apply

```python
data['valor_m2']= data.anuncio_valores_venda / data.imovel_area
```

```python
data[['anuncio_valores_venda','imovel_area', 'valor_m2']].iloc[16:19]
```

```python
valor_m2 = lambda x: round(x['anuncio_valores_venda'] / x['imovel_area'],2) if x['imovel_area'] != 0 else 0
```

```python
data['valor_m2'] =  data.apply(valor_m2, axis = 1)
```
