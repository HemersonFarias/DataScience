```python
dados = pd.read_json("https://raw.githubusercontent.com/sthemonica/alura-voz/main/Dados/Telco-Customer-Churn.json",orient = 'columns')
dados.head(10)
dados.rename(columns = {'customerID': 'ID_Cliente'}, inplace = True)
dados
```

```python
dados2 = pd.read_json("https://raw.githubusercontent.com/sthemonica/alura-voz/main/Dados/Telco-Customer-Churn.json",orient = 'columns')
dados2.head(10)
```

```python
dados_customer = pd.json_normalize(data = dados.customer)
dados_customer.columns = ['Cliente_Genero', '>65 anos', 'Cliente_com_parceiro', 'Cliente_dependentes','Cliente_meses_de_contrato']

dados_phone = pd.json_normalize(data = dados.phone)
dados_phone.columns = ['Assinatura_Servico_Telefonico', 'Assinatura_mais_de_1_linha_Telefone']

dados_internet = pd.json_normalize(data = dados.internet)
dados_internet.columns = ['Internet_Provedor', 'Internet_seguranca_online', 
                          'Internet_Backup_Online', 'Internet_protecao_dispositivo',
                          'Internet_suporte', 'Internet_TV_cabo', 'Internet_filmes_straming']

dados_account = pd.json_normalize(data = dados.account)
dados_account.columns = ['Tipo_contrato','Fatura_Online?','Metodo_Pagamento', 'Total_servicos_cliente_mes', 'Total_gasto_Mes']
```

```python
dados = pd.concat([dados.ID_Cliente, dados.Churn ,dados_customer, dados_phone, dados_internet, dados_account], axis = 1)
```
